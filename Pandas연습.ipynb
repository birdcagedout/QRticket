```python
import pandas as pd
```

```python
# Dict로 df만들기

friend_list = {
	'name': ['john', 'jenny', 'nate'],
	'age': [20, 30, 40],
	'job': ['student', 'developer', 'teacher']
}

df = pd.DataFrame(friend_list)
```

```python
df
```

```python
df[1:3]
```

```python
len(df[1:3])    # row의 개수
```

```python
df.loc[[0,2]]    # 특정 row만 선택(row index)
```

```python
df[df.age >= 30]    # 특정 row만 선택(col값 조건)
```

```python
df.query("age >= 30")
```

```python
df[ (df.age >= 30) & (df.name == 'nate')]
```

### Filter Columns
#### 1) index

```python
# list로 df 만들기

friend_list = [
	['john', 20, 'student'],
	['jenny', 30, 'developer'],
	['nate', 40, 'teacher']
]

df = pd.DataFrame.from_records(friend_list)
df
```

```python
df.columns = ['name', 'age', 'job']
df
```

```python
df.iloc[:, 0:2]    # row는 전부(:), col은 0,1번째만
```
